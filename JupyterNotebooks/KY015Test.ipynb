```python
import time
from pynq.overlays.base import BaseOverlay
from pynq.lib.arduino.arduino_io import Arduino_IO
base = BaseOverlay("base.bit")
```

```python
PIN = 8
data = [0, 0, 0, 0, 0]
USCALE = 1000000
arduino_pin = Arduino_IO(base.ARDUINO, 8, 'out') #see pin AR8 on pynq board
LOW, HIGH = 0, 1 #https://www.arduino.cc/reference/en/language/variables/constants/constants/
```

```python
def read_data():
    print("reading data...")
    i = 0b0000000 #byte i = 0
    result = 0b00000000#byte result = 0
    for i in range(8):
        while arduino_pin.read()==LOW: 
            pass #wait for arduino pin to be low
        time.sleep(30/USCALE) #wait for 30us
        if arduno_pin.read()==HIGH:
            result |= (0b1<<(0b1000-i))
        while arduino_pin.read()==HIGH:
            pass #wait for pin to be hight 
    return result
```

```python
def start_test():
    print("starting test...")
    arduino_pin.write(LOW)
    time.sleep(0.03)#wait for 30ms
    arduino_pin.write(HIGH)
    time.sleep(40/USCALE)
    arduino_pin.direction = 'in'
    while arduino_pin.read()==HIGH:
        pass
    time.sleep(80/USCALE)
    if arduino_pin.read()==LOW:
        time.sleep(80/USCALE)
    for i in range(5):
        data[i] = read_data()
    arduino_pin.direction = 'out'
    arduino_pin.write(HIGH)    
```

```python
def setup():
    arduino_pin.direction = 'out'
```

```python
def loop():
    print("starting loop...")
    start_test()
    print(data)
```

```python
for i in range(10):
    setup()
    loop()
```
